```python
## Use this cell to define the function corresponding to OPTION 1

def option1():
    
    '''With this function you input name and id number.'''
    
    global name
    print("Please enter your name:")
    name = input().capitalize()
    print("Please enter your id number:")

    while True:
        try:
            id_number = input()
            number_list = list(map(int, list(id_number)))
            print("Hi " + name + ", the sum of the digits of your id is: " + str(sum(number_list)))
            break
        except ValueError:
            print("Invalid input. Please enter your id number:")
        
    return 
```

```python
## Use this cell to define the function corresponding to OPTION 2

def option2():
    
    '''This function prints the number of "a's" in the name.'''
    
    count_name = name.count('a')
    print('The number of letter "a" in your name is: ' + str(count_name))
    
    
    return
```

```python
## Use this cell to define the function corresponding to OPTION 3

def option3():
    '''This function allows the user to create the students and marks lists.'''
    
    while True:
        
        try:
            print("Indicate the number of students/marks that you want to include: ")
            z = int(input())
            break
            
        except ValueError:
            print("Please, enter a number.")
            
    global students
    students = []
    for i in range(0,z):
        while z != len(students):
            print("Please enter a name:")
            student_input = input().capitalize() 

            if not student_input in students:
                students.append(student_input)
            else:
                print('Name already in use. Please try again.')
    
    global marks
    marks = []

    for pupil in students:
        while True:
            try:
                mark_input = float(input('Please enter the mark for ' + pupil + ' (between 0 and 10): '))
                
                if round(mark_input) in range(0,11):
                    marks.append(mark_input)
                    break
            
                else:
                    print('Invalid mark. Please try again.')
                
            except ValueError:
                print("Invalid mark. Please enter a number between 0 and 10")
    
    
    return
```

```python
## Use this cell to define the function corresponding to OPTION 4

def option4():
    
    '''This function allows to search and print the mark of a student.'''
    
    trials = 3
    while trials > 0:
    
        print("Please enter a student's name: ")
        s = input().capitalize()

        if s in students:
            p = students.index(s)
            print ("The mark of " + s + " is " + str(marks[p]))
            break

        else: 
            print("Name not found. Please try again.")     
            trials -= 1
    
        if trials == 0:     
            print("Name not found for the third time. Exiting the function.")
            break    

        
    return
```

2. In option 3 you had to create two lists, one for the students and one for the marks. Is there any other data structure in Python that would help you reduce the number of variables and that can facilitate the way to access the marks?.

**ANSWER:**

The use of a dictionary with keys being the names of students and values being their marks should facilitate the way to access the marks: instead of retrieving an element through its position in a list, with dictionaries it is possible to access an element through its key (student's name in this case), eliminating the possibility of a mistake with the position in the list.
It would also reduce the number of variables (from 2 to 1) and even ensure that no names are entered twice - since it does not support duplicates - without additional programming lines

```python
# an example:

# asking how many number to include in the dictionary

while True:
    try:
        print("Indicate the number of students/marks that you want to include: ")
        z = int(input())
        break
        
    except ValueError:
        print("Please, enter a number.")
    

students_marks = {}  # creating an empty dictionary
for i in range(0,z):
    while z != len(students_marks.keys()):
        print("Please enter a name:")
        student = input().capitalize() 
        
        students_marks[student] = 0 # inserting the user inputs as keys in the students_marks dictionary 
                                    # and assigning them a starter value of 0

# if we did not use a dictionary, we would need to make sure that the user does not enter the same name twice

        # IF NOT student IN students_marks.keys():
        #   students_marks[student] = 0
        # ELSE:
            # print('Name already in use. Please try again.')

# however, since dictionaries do not support duplicate, these lines are redundant and unnecessary

        
for key in students_marks:
    while True:
        try:
            mark = float(input('Please enter the mark for ' + key + ' (between 0 and 10): '))
                
            if round(mark) in range(0,11):
                students_marks[key] = mark # substituting the starter value of 0 with user's inputs for each 
                break                      # key/student
            
            else:
                print('Invalid mark. Please try again.')
                
        except ValueError:
            print("Invalid mark. Please enter a number between 0 and 10")
                

print(students_marks)

# then, to retrieve the mark of a student

y = input("Please enter the name of the student to retrieve his/her mark: ").capitalize()
print("The mark of " + y + " is " + str(students_marks[y]))
```
