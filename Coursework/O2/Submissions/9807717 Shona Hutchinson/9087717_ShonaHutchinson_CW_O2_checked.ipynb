**ANSWER:**

The polyfit takes the the x and y data as input and returns the constant and the slope. (m & b in the equation y = mx + b)
R- Squared (the coefficient of determination ) can be used to decide which 2 variables are most correlated.  The value of R-Squared ranges from 0 to 1.  The closer to 1 the better the regression line equation explains the relationship between the 2 variables.  If R-Squared was exactly 1 then for all the observed data, the dependent variable y always sits exactly on the regression line.

The pearsonr class returns the person r coefficient which is the square root of R Squared.
There is not much of a correlation between any of the 3 variables but the strongest is between latitude and longitude which has a weak negative relationship. (R Squared is 0.29) 

At the end of the code cell below it will ask for an x value for which to predict a value for y using the equation of the regression line.  However as the R Squared vales are so weak it's not actually worth bothering with.
